```python
import spacy
```

```python
nlp = spacy.load('en_core_web_sm') # English dictionary
```

```python
doc = nlp(u"I will Google about Facebook")
```

```python
doc.text
```

```python
doc[0] # you can use this do as array to indexing them ad get the proper words
```

```python
doc[1].pos_ # You getting AUX is called as Core Grain Part of speech
```

```python
doc[2].tag_
```

```python
spacy.explain('NNP') # out -> noun, proper singular so this proper singular is additional detailing & it is called Fine Grain Parts of Speech
```

```python
for word in doc:
  print(word.text, "----->", word.pos_, word.tag_, spacy.explain(word.tag_))
```

```python
# Word sense Ambiguition
doc2 = nlp(u"I left the room")
for word in doc2:
  print(word.text, "----->", word.pos_, word.tag_, spacy.explain(word.tag_))
```

```python
doc3 = nlp(u"to the left of the room")
for word in doc3:
  print(word.text, "----->", word.pos_, word.tag_, spacy.explain(word.tag_))

# See two left has different meaning so their pos is also different
```

```python
# Grammatical Handling
doc4 = nlp(u"I read books on history")
for word in doc4:
  print(word.text, "----->", word.pos_, word.tag_, spacy.explain(word.tag_))
```

```python
doc5 = nlp(u"I have read books on history")
for word in doc5:
  print(word.text, "----->", word.pos_, word.tag_, spacy.explain(word.tag_))

# We just add have in the sentence to convert the sentence in present perfect so above where read as singular present but
# here read is as past participle (3rd form of verb)
```
