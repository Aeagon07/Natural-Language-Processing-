```python
import gensim
import os
```

```python
from nltk.tokenize import sent_tokenize
from gensim.utils import simple_preprocess
import nltk
nltk.download('punkt_tab')

story = []
for filename in os.listdir('/content/data'):
  # Attempt to open with 'latin-1' encoding to handle potential character issues
  f = open(os.path.join('/content/data', filename), encoding='latin-1')
  corpus = f.read()
  raw_sent = sent_tokenize(corpus)
  for sent in raw_sent:
    story.append(simple_preprocess(sent))

# Here you did the simple preproccessing on the each book separated by the sentances and append them into story array
```

```python
len(story) # here you get the sentances and tokenize them into words
```

```python
# create your model
model = gensim.models.Word2Vec(
    window=10, # center word ke aage piche 10-10 words honge
    min_count=2
)
```

```python
# Buid Vocab
model.build_vocab(story)
```

```python
model.train(story, total_examples=model.corpus_count, epochs=model.epochs) # by default it train on the default 5
```

```python
try:
    print(model.wv.most_similar('daenerys'))
except KeyError:
    print("Key not found")
```
