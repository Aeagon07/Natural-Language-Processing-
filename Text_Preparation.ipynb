```python
# this function will help you out to remove the tags in the text
import re
def remove_html(text):
  pattern = re.compile('<[^>]+>')
  return pattern.sub(r'', text)
```

### Removal of Punctuation

```python
import string, time
string.punctuation
```

```python
exclude = string.punctuation
```

```python
def remove_puc(text):
  for ch in exclude:
    text = text.replace(ch, '')
  return text

def remove_puc1(text):
  return text.translate(str.maketrans('', '', exclude))
```

```python
new_text = 'String, with Punctuation?'
```

```python
remove_puc1(new_text)
```

### Chat Word Treatment

```python
chat_shortcuts = {
    "u": "you",
    "ur": "your",
    "urs": "yours",
    "r": "are",
    "btw": "by the way",
    "idk": "i do not know",
    "imo": "in my opinion",
    "imho": "in my humble opinion",
    "brb": "be right back",
    "bbl": "be back later",
    "lol": "laughing out loud",
    "lmao": "laughing my ass off",
    "rofl": "rolling on the floor laughing",
    "omg": "oh my god",
    "ttyl": "talk to you later",
    "pls": "please",
    "plz": "please",
    "thx": "thanks",
    "ty": "thank you",
    "np": "no problem",
    "gr8": "great",
    "b4": "before",
    "afk": "away from keyboard",
    "irl": "in real life",
    "msg": "message",
    "dm": "direct message",
    "jk": "just kidding",
    "cuz": "because",
    "bc": "because",
    "w8": "wait",
    "ya": "you",
    "yeah": "yes",
    "nah": "no",
    "ok": "okay",
    "k": "okay",
    "ttys": "talk to you soon",
    "smh": "shaking my head",
    "tbh": "to be honest",
    "rn": "right now",
    "fr": "for real",
    "wyd": "what are you doing",
    "wbu": "what about you",
    "hru": "how are you",
    "gtg": "got to go",
    "atm": "at the moment",
    "fyi": "for your information"
}
```

```python
chat_shortcuts.pop('wyd')
```

```python
# Create the Function
def chat_convertor(text):
  new_text = []
  for w in text.split():
    if w.lower() in chat_shortcuts:
      new_text.append(chat_shortcuts[w.lower()])
    else:
      new_text.append(w)
  return " ".join(new_text)
```

### Removing Stop Words

```python
from nltk.corpus import stopwords
```

```python
import nltk
nltk.download('stopwords')
```

```python
stopwords.words('english')
```

```python
def remove_stWord(text):
  new_text = []

  for word in text.split():
    if word is stopwords.words('english'):
      new_text.append('')
    else:
      new_text.append(word)
  x = new_text[:]
  new_text.clear()
  return " ".join(x)
```

```python
example ='i do not know what you are doing right now laughing out loud'

remove_stWord(example)
```

### Remove Emoji

```python
import re
def remove_emoji(text):
  emoji_pattern = re.compile("["
        "\U0001F600-\U0001F64F"  # emoticons
        "\U0001F300-\U0001F5FF"  # symbols & pictographs
        "\U0001F680-\U0001F6FF"  # transport & map symbols
        "\U0001F700-\U0001F77F"
        "\U0001F780-\U0001F7FF"
        "\U0001F800-\U0001F8FF"
        "\U0001F900-\U0001F9FF"  # supplemental symbols
        "\U0001FA00-\U0001FAFF"  # extended symbols
        "\u2600-\u26FF"          # misc symbols
        "\u2700-\u27BF"          # dingbats
        "]+", flags=re.UNICODE)

  return emoji_pattern.sub(r'', text)
```

```python
remove_emoji(text7)
```

# **Stemming**

```python
# Using NLTK's Portor Stemmers -> Used only for English Langauge convertion
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
def steam_word(text):
  return " ".join([ps.stem(word) for word in text.split()])
```

```python
sample = "walk walks walking walked"
steam_word(sample)
```

# **Lemmatization**

```python
import nltk
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
wordnet_lemmatizer = WordNetLemmatizer()

sentence = "He was running and eating at same time. He has bad habbit of swimming after playing long hours in the Sun."
punct = "?:!.,;"
sent_word = nltk.word_tokenize(sentence)

# Remove all Punctuation
for word in sent_word:
  if word in punct:
    sent_word.remove(word)

sent_word
print("{0:20}{1:20}".format("Word", "Lemma"))
for word in sent_word:
  print("{0:20}{1:20}".format(word, wordnet_lemmatizer.lemmatize(word, pos='v'))) # you have to specify the parts of speech
```
