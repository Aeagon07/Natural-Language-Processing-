# **Bag Of Word**

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'text': ['people watch campusx', 'campusx wath campusx', 'people write comment', 'campusx write comment'], 'output': [1, 1, 0, 0]})
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```python
bow = cv.fit_transform(df['text'])
```

```python
# Vocab
print(cv.vocabulary_)
# the word is alphabetically arranged so they give them number so when your array gives you [1, 0, 0, 0, 0] that means there is only campusx which is present
```

```python
print(bow[0].toarray())
print(bow[1].toarray())
print(bow[2].toarray())
```

```python
cv.transform(["campusx watch and write comment of campusx"]).toarray()
```

# **N-Grams**

```python
#  This is same as the bag of words cause BoW is also a Uni-Gram special technique of N-Grams
df
```

```python
cv2 = CountVectorizer(ngram_range=(2, 2)) # for Bi-Grams {default ngram_range is set to (1, 1)}
```

```python
bow = cv2.fit_transform(df['text'])
```

```python
print(cv2.vocabulary_)
```

```python
print(bow[0].toarray()) # D1
```

# **Tf-Idf**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
tfidf.fit_transform(df['text']).toarray()
```

```python
print(tfidf.idf_)
print(tfidf.get_feature_names_out())

# Why these values are higher => they have very different technique that after log they added (+1) so they avoid those terms which are occurs in all the corpus
# and so it's log value become 0 and after multiplication with the tf it will give you 0 so that...
```
