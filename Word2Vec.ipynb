```python
model = KeyedVectors.load_word2vec_format(
    "GoogleNews-vectors-negative300.bin",
    binary=True
)
```

```python
model['man']
# From this you will understand that they are dense vectors and the numbers are non-zero in nature
```

```python
model['man'].shape
```

```python
model.most_similar('man')
```

```python
# You can also find the cosine similarity between the two vectors
model.similarity('man', 'woman')
```

```python
model.similarity('man', 'PHP')
```

```python
# You can find least similar in the given array i.e., odd one out
model.doesnt_match(['PHP', 'java', 'monkey'])
```

```python
# You can do vector arithmatic also
vec = model['king'] - model['man'] + model['woman']
model.most_similar([vec])
```

```python
vec = model['messi'] - model['football']
model.most_similar([vec])
```
